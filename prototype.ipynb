```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
```

```python
import openpyxl as xl 
```

```python
import seaborn as sns
```

```python
import os
from pathlib import Path
```

```python
root = Path.cwd()
file_path = root / Path("Auswertung von Teilnehmer- und Dozentenfeedback_MDZ.xlsm")

file_path
```

```python
df = pd.read_excel(file_path, sheet_name="Auswertungsblatt (QF)", engine="openpyxl", header=1)
```

```python
df.dropna(subset=['Datum'],inplace=True)
```

```python
df
```

```python
df.columns
```

```python
df_tn_1 = df.iloc[:, :14]
```

```python
df_tn_1
```

```python
df_tn_1.set_index('Datum', inplace=True)
```

```python
plt.figure(figsize=(60,25))
```

```python
auswertungs_col = list(df_tn_1.columns)[5:]
```

```python

sns.lineplot(data=df_tn_1.loc[:,auswertungs_col])
plt.axhspan(1, 2.5, facecolor='lightgreen', alpha=0.5)
plt.axhspan(2.5, 3.3, facecolor='yellow', alpha=0.5)
plt.axhspan(3.3, 5, facecolor='red', alpha=0.5)
ax= plt.gca()
ax.axhline(y=3.3, color='red', linestyle='--', label='Dozenten Kontaktieren')
plt.xticks(rotation=45)
plt.xlabel('Zeitachse')
plt.ylabel('Teilnehmerbewertung')
plt.title('Evolution der Teilnehmerbewertung im MDZ')
#legend_patch = Rectangle((0, 0), 1, 1, facecolor='green', alpha=0.3)
#plt.legend([legend_patch], ['Shaded Region'], loc='upper right')
plt.legend(loc='upper right', bbox_to_anchor=(1.6,1))
```

```python

untersucht=auswertungs_col[0] # Change index here to get other measures
sns.lineplot(data=df_tn_1.loc[:,untersucht])
plt.axhspan(1, 2.5, facecolor='lightgreen', alpha=0.5)
plt.axhspan(2.5, 3.3, facecolor='yellow', alpha=0.5)
plt.axhspan(3.3, 5, facecolor='red', alpha=0.5)
ax= plt.gca()
ax.axhline(y=3.3, color='red', linestyle='--', label='Dozenten Kontaktieren')
plt.xticks(rotation=45)
plt.xlabel('Zeitachse')
plt.ylabel('Teilnehmerbewertung')
plt.title(f'Evolution der {untersucht[2:]} im MDZ')
#legend_patch = Rectangle((0, 0), 1, 1, facecolor='green', alpha=0.3)
#plt.legend([legend_patch], ['Shaded Region'], loc='upper right')
plt.ylim(1,3.5)
plt.legend(loc='upper right', bbox_to_anchor=(1.45,1))
```
